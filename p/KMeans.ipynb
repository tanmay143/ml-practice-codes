# importing the libraries


```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

# reading the dataset

```python
df = pd.read_csv('wine.data')
df.head()
```

# K Means Model

```python
km=KMeans(n_clusters=3, max_iter=400, verbose=True, tol=0.2)
```

```python
pred = km.fit_predict(df[['14.23', '1.71']])
```

# scatter plot before k means 

```python
plt.scatter(df['14.23'], df['1.71'], c=df['15.6'])
```

# scatter plot after k means 

```python
plt.scatter(df['14.23'], df['1.71'], c=pred)
```

# Accuracy score value

```python
score = accuracy_score(df['1'], pred)
score
```
