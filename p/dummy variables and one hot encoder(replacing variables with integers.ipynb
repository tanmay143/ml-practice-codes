```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('book1.csv')

np.shape(df)
```

```python
df.head(5)
```

```python
dummies = pd.get_dummies(df[['Car Model']])
dummies
```

```python
merge = pd.concat([df,dummies],axis='columns')
merge
```

```python
final = merge.drop(['Car Model','Car Model_Mercedez Benz C class','Sell Price($)'], axis ='columns')
y=df[['Sell Price($)']]
y
```

```python
reg = LinearRegression()
reg.fit(final,y)
```

```python
final
```

```python
reg.score(final,y)
```
